```python
import requests
from bs4 import BeautifulSoup
import urllib.parse
import time

def read_urls_from_file(filename):
    with open(filename, 'r') as file:
        return [line.strip() for line in file if line.strip()]

def get_restaurant_address(url):
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        
        address_div = soup.find('div', class_='flex flex-row items-center justify-between px-1 w-full md:w-fit')
        if address_div:
            address_parts = address_div.find_all('p')
            if len(address_parts) >= 2:
                return f"{address_parts[0].text.strip()} {address_parts[1].text.strip()}"
        
        return "Address not found"
    except requests.RequestException:
        return "Error fetching address"

def write_to_js(restaurants):
    with open('restaurants.js', 'w') as f:
        f.write('const restaurants = [\n')
        for restaurant in restaurants:
            f.write(f'    {{ name: "{restaurant["name"]}", address: "{restaurant["address"]}" }},\n')
        f.write('];')

urls_file = 'url.text'
restaurant_links = read_urls_from_file(urls_file)
addresses = []
for link in restaurant_links:


    address = get_restaurant_address(link)
    if address != "Address not found" and address != "Error fetching address":
        addresses.append(address)
    print(f"Restaurant: {link}")
    print(f"Address: {address}")
    print("-" * 50)
    time.sleep(0.1)  # Adding a delay to be respectful to the server
    
google_maps_link = create_google_maps_link(addresses)
print("\nGoogle Maps link with all locations:")
print(google_maps_link)
write_to_js(addresses)
```

```python
def create_google_maps_link(addresses):
    base_url = "https://www.google.com/maps/search/?api=1&query="
    encoded_addresses = urllib.parse.quote("|".join(addresses))
    return base_url + encoded_addresses
```

```python
import time
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut, GeocoderServiceError

geolocator = Nominatim(user_agent="your_app_name (your_email@example.com)")

def clean_address(address):
    return address.replace(", Vienna, Austria", "").strip()

def geocode_address(address, retries=3):
    cleaned_address = clean_address(address)
    for attempt in range(retries):
        try:
            print(f"Attempting to geocode: {cleaned_address} (Attempt {attempt + 1}/{retries})")
            location = geolocator.geocode(cleaned_address)
            if location:
                return (location.latitude, location.longitude)
            else:
                print(f"No results found for: {cleaned_address}")
                return None
        except (GeocoderTimedOut, GeocoderServiceError) as e:
            print(f"Error occurred: {e}. Retrying in 1 second...")
            time.sleep(1)  # Wait for 1 second before retrying
    print(f"All retries failed for: {cleaned_address}")
    return None  # Return None if all retries fail


# Process each address
for address in addresses:
    print(f"\nProcessing address: {address}")
    geo_point = geocode_address(address)
    print(f"Original Address: {address}")
    print(f"Cleaned Address: {clean_address(address)}")
    print(f"Geo Point: {geo_point}\n")
    time.sleep(1)  # Wait for 1 second between requests to respect rate limits
```

```python
import json
from bs4 import BeautifulSoup

def parse_restaurant_html(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    
    restaurant_data = {}
    
    # Extract restaurant name
    name_element = soup.find('h1', class_='text-[40px] md:text-[36px] font-bold md:font-semibold')
    if name_element:
        restaurant_data['name'] = name_element.text.strip()
    
    # Extract rating
    rating_element = soup.find('span', class_='relative font-medium text-[12px] md:text-[16px] text-black/60')
    if rating_element:
        restaurant_data['rating'] = rating_element.text.strip()
    
    # Extract number of reviews
    reviews_element = soup.find('div', class_='font-medium text-[12px] md:text-[16px] text-black/60')
    if reviews_element:
        reviews_text = reviews_element.text.strip()
        restaurant_data['reviews'] = reviews_text.split()[0]  # Get the first word (number)
    
    # Extract categories
    categories_element = soup.find('p', class_='w-full max-w-fit whitespace-nowrap overflow-hidden text-ellipsis')
    if categories_element:
        restaurant_data['categories'] = categories_element.text.strip()
    
    # Extract price range
    price_range_elements = soup.find_all('span', class_='text-black/80')
    restaurant_data['price_range'] = len(price_range_elements)
    
    return restaurant_data
```
